```python
df.describe(include=['float', 'int'])
```

```python
df_1 = df.copy(deep = True) #создаем копию датасета
median_Age = df_1.Age.median() # используется стратегия замены медианным значением

df_1 = df_1.fillna({'Cabin':'A0', 'Embarked':'A', 'Age':median_Age}) # т.к. данных по каюте и порту посадки у нас по некоторым пассажирам нет, а вычислить среднее или медианное значение невозможно, принято решение подставить по ним константы, отличающиеся от иных значений
df_1
```

```python
df_1.isnull().sum()
```

```python
# Альтернативный вариант

from sklearn.impute import SimpleImputer  #импортируем библиотеку
df_2 = df.copy(deep = True) #создаем копию датасета

myImputer_median = SimpleImputer (strategy= 'median') # используется стратегия замены медианным значением

myImputer_mf = SimpleImputer (strategy= 'most_frequent') # используется стратегия замены наиболее часто встречающимся значением

#используем метод fit_transform для замены пропущенных значений

df_2[['Cabin','Embarked']]=myImputer_mf.fit_transform(df_2[['Cabin','Embarked']])
df_2.Age = myImputer_median.fit_transform(df_2['Age'].values.reshape(-1,1))
df_2
```

# Пункт 8

```python
df_part_SP = df_1[['Pclass', 'Survived']]
survived_pclass_rate = df_part_SP.groupby('Pclass').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1]), len(x)))


for i, rate in survived_pclass_rate.items():
  print(f'с классом круиза {i} доля выживших: {rate[0]}, общее число выживших в данном классе: {rate[1]}, общее число пассажиров в данном классе: {rate[2]}')
```

Вывод: чем выше класс, тем больше было шансов выжить

```python
df_part_SA = df_1[['Age', 'Survived']]
survived_age_rate = df_part_SA.groupby('Age').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1]), len(x)))


for i, rate in survived_age_rate.items():
  print(f'в возрасте {i} доля выживших: {rate[0]}, общее число выживших в данном возрасте: {rate[1]}, общее число пассажиров в данном возрасте: {rate[2]}')
```

Вывод: Слишком много значений возрастов, необходимо сгруппировать людей по диапазонам возрастов

```python
# Создадим список с диапазонами возрастов, где значение = предельному значению возраста в диапазоне
Age_range = []
for row in df_1['Age']:
        if row <= 20.0 :    Age_range.append(20)
        elif row <= 40.0:   Age_range.append(40)
        elif row <= 60.0:  Age_range.append(60)
        elif row <= 80.0:  Age_range.append(80)

        else:           Age_range.append('Not_Rated')
```

```python
# Добавим список как столбейц в таблицу
df_1['Age_range'] = Age_range
df_1
```

```python
df_part_SAR = df_1[['Age_range', 'Survived']]
survived_age_range_rate = df_part_SAR.groupby('Age_range').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1]), len(x)))


for i, rate in survived_age_range_rate.items():
  print(f'в возрасте до {i} лет доля выживших: {rate[0]}, общее число выживших в данном диапазоне: {rate[1]}, общее число пассажиров в данном диапазоне: {rate[2]}')
```

Вывод: Люди до 60 лет имели больше шансов выжить

```python
df_part_SS = df_1[['Sex', 'Survived']]
survived_sex_rate = df_part_SS.groupby('Sex').apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1]), len(x)))


for i, rate in survived_sex_rate.items():
  print(f'доля выживших {i}: {rate[0]}, общее число выживших данного пола: {rate[1]}, общее число пассажиров данного пола: {rate[2]}')
```

Вывод: У женщин шанс выживания был практически в 4 раза выше

```python
df_part_SSP = df_1[['SibSp', 'Parch', 'Survived']]
survived_relatives_rate = df_part_SSP.groupby(['SibSp', 'Parch']).apply(lambda x: (len(x[x['Survived'] == 1])/len(x), len(x[x['Survived'] == 1]), len(x)))


for i, rate in survived_relatives_rate.items():
  print(f'доля выживших с родственниками в сестер/братьев/супругов {i[0]} и  родителей/детей {i[1]} человек на борту: {rate[0]}, общее число выживших с данным количеством родни на борту: {rate[1]}, общее число пассажиров с данным количеством родни на борту: {rate[2]}')
```

Вывод: практически во всех комбинациях, наличие родственников на борту прибавляло процент шанса выживания
