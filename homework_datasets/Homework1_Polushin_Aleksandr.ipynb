# Решение с помощью библиотеки NumPy

```python
#импорт библиотеки
import numpy as np
```

```python
#перемножаемые матрицы
A=np.matrix([[6, -8, 3]])
B=np.matrix([[1],
             [1],
             [4]])

#вывести результат умножения
print(A * B)
```

```python
#перемножаемые матрицы
C=np.matrix([[-3, -7],
             [-10, 6],
             [0, -6],
             [-1, -3]])
D=np.matrix([[-8, 9],
             [4, -2]])

#вывести результат умножения
print(C * D)
```

# Решение через циклы

```python
def matmul(matrix1_,matrix2_):
    result = [] #финальный результат

    for i in range(len(matrix1_)):

        row = [] #строка в матрице
        for j in range(len(matrix2_[0])):

            product = 0 #элемент в строке

            for v in range(len(matrix1_[i])):
                 product += matrix1_[i][v] * matrix2_[v][j]
            row.append(product) #добавить сумму произведения в новую строку

        result.append(row) #добавить новую строку в финальный результат
    return result
```

```python
#перемножаемые матрицы
A=([[6, -8, 3]])
B=([[1],
    [1],
    [4]])

#выполнить умножение матриц
result = matmul(A, B)

#вывести результат умножения
print(result)
```

```python
#перемножаемые матрицы
C=([[-3, -7],
    [-10, 6],
    [0, -6],
    [-1, -3]])
D=([[-8, 9],
    [4, -2]])

#выполнить умножение матриц
result = matmul(C, D)

#вывести результат умножения
print(result)
```
