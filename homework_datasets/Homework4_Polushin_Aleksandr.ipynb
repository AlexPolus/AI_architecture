```python
import numpy as np

np.__version__
```

```python
#create matrix A
A = np.matrix('1 2 1;-2 1 1')
print('матрица А')
print(A)

#transpose matrix A
AT = np.asarray(A).transpose()
print('транспонированная матрица А')
print(AT)

#create matrix B
B = np.matrix('3 1;-1 0;2 4')
print('матрица B')
print(B)

#create matrix C
C = np.matrix('-2 2;1 -1;1 2')
print('матрица C')
print(C)

print('solution A+2B')
print(AT+2*B)

print('solution B+2C')
print(B+2*C)
```
